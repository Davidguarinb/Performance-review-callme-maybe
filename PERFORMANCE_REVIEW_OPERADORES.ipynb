# **Paso 4**

Una vez hecho el trabajo de exploración y procesamiento de datos, se analiza y se propone una solución concreta para el problema de la compañía.

Se desea conocer qué operadores son inficaces, para la compañía un operador es ineficaz si:
  - Número de llamadas entrantes perdidas alto (internas y externas)
  - Un Número reducido de llamadas salientes.
  - Un tiempo de espera prolongado

Como no hay valores previos de indicadores base se tomará la media de cada valor como una forma de medir. En este caso los sesgos de los datos son tan favorables como desfavorables (para empleados y empresa) para medir el rendimmiento.

Así que:
  1. Se propone una agrupación por operator_id
  2. Se contará la media de llamadas perdidas, número de llamadas salientes y tiempo de espera por operator_id
  3. Se crea la primera condición: si un operador está por encima de la media  de llamadas entrantes perdidas se considera ineficaz.
  4. Se crea la segunda condición: si un operador está por debajo de la media de llamadas salientes se considera ineficaz.
  5. Se crea la tercera condición: si un operador está por encima de la media del tiempo de espera se considera ineficaz.

Posteriormente se formulará una hipótesis y se realizará su prueba.

*Como parte del primer paso se comproborá si existe presencia de operator_id en el user_id.

````python
# verificación presencia operator_id en user_id
id_operadores = eventos_copy['operator_id'].unique() #Filtro que busca el listado de id unico para cada operador
id_usuarios = eventos_copy['user_id'].astype('str').unique() #Filtro que busca el id unico para cada usuario

operadores_en_usuarios = set(id_operadores) & set(id_usuarios) # Operadores para buscar si algún id_de operador se encuentra en la lista de usuarios
operadores_no_en_usuarios = set(id_operadores) - set(id_usuarios) #Operadores para puscar si algún usuario se encuentra en la lista de operadores
````

Al revisar este filtro, es obvio que las llamadas de internal, no son de operador a operador y definitivamente se deja la definición inicial de: usuarios de la misma red. Aunque en las indicaciones del cliente se mencionó que valián tanto llamadas internas, como externas, igual se hizo esta revisión para evitar posibles conflictos a la hora de realizar las agrupaciones que responderán que operador es ineficaz.

````python
#Calculo y creación de la columna de tiempo de espera
eventos_copy['tiempo_espera'] = eventos_copy['total_call_duration']-eventos_copy['call_duration']

#Creación de dataframes filtrados por llamadas entrantes y salientes.
df_in  =   eventos_copy[eventos_copy['direction'] == 'in'].copy()
df_out = eventos_copy[eventos_copy['direction'] == 'out'].copy()
````

Se crearon dataframes diferentes para el filtro de llamadas entrantes y llamadas salientes, debido a que las condiciones para ambas llamadas con distintas. Las llamadas entrantes dependen de la agilidad del operador para responderlas, las llamadas salientes dependen de la disposición del usuario para responder, en cuyo caso solo se evaluará el número de duración de las llamadas.

````python
#Se calcula la métrica de llamadas perdidas, duración de llamadas  y tiempo de espera para las llamadas entrantes por operador
metricas_in = (
    df_in
    .groupby('operator_id')
    .agg(
          missed_call_rate = ('is_missed_call', 'mean'),
          ingoing_calls_mean =('calls_count', 'mean'),
          wait_time_mean = ('tiempo_espera','mean')
    )
    .reset_index()
)

#Se calcula la métrica de duración de las llamadas para las salientes

metricas_out = (
    df_out
    .groupby('operator_id')
    .agg(
          outgoing_calls_mean = ('calls_count', 'mean')
    )
    .reset_index()
)

#Dataframe performance por operador
#Creación de dataframe con los id's unicos de cada operador
unique_operator_ids = eventos_copy['operator_id'].unique()
operator_id_df = pd.DataFrame(unique_operator_ids, columns=['operator_id'])


#Se une el dataframe recién creado con los id que aparecen  en metricas_operador['operator_id_in'] que corresponde al performance de los operadores que respondieron llamadas entrantes y sus métricas repectivas
merged_metrics_1 = pd.merge(operator_id_df, metricas_in, left_on='operator_id', right_on='operator_id', how='left')
#merged_metrics_1 = merged_metrics_1.drop(columns=['index', 'operator_id'])

#Se une el dataframe recién creado con los id que aparecen  en metricas_operador['operator_id_in'] que corresponde al performance de los operadores que respondieron llamadas entrantes y sus métricas repectivas con la sección de id's de metricas_operador['operator_id_out'] que corresponde a las métricas de las llamadas salientes
merged_metrics_2 = pd.merge(operator_id_df, metricas_out, left_on='operator_id', right_on='operator_id', how='left')

#Se unen los dos Dataframes filtrados por las métricas clave

metricas_operador = merged_metrics_1.merge( merged_metrics_2)


metricas_operador
````

````python
#Valores medios promedio por operador
media_missed = metricas_operador['missed_call_rate'].mean()
media_ingoing = metricas_operador['ingoing_calls_mean'].mean()
media_outgoing = metricas_operador['outgoing_calls_mean'].mean()
media_wait_time = metricas_operador['wait_time_mean'].mean()
print('Valor medio de llamadas perdidas por operador:', media_missed)
print('Valor medio de duración de cada llamada entrante por operador:', media_ingoing)
print('Valor medio de número llamadas salientes por operador:',media_outgoing)
print('Valor medio de duración del tiempo de espera para cada llamada entrante por operador:',media_wait_time)
````

Los valores medios de las llamadas entrantes (0 rechazos por cada operador, 4.1 segundos de duración para cada llamada y tiempo de espera de 57 segundos) marcan un condicional, el valor medio de las llamadas salientes (14 segundos de duración de las llamadas) marca el segundo condicional. Se procede a construir una función que itere sobre cada línea y califique si los operadores son inficaces, por cada llamada. Si clasificará a cada uno de los 1093 operadores en función del valor que más se repita según el performance de cada llamada.

````python
#Creación de filtros para cada

#Condición 1: Evaluación del performance con respecto a las llamadas perdidas
metricas_operador['ineficaz_perdidas'] = (metricas_operador['missed_call_rate'] > media_missed)

#condición 2: Evaluación del performance con respecto al tiempo de espera para recibir la llamada
metricas_operador['ineficaz_tiempo_espera'] = (metricas_operador['wait_time_mean'] < media_outgoing)

#condición 3: Evaluación del performance con respecto al tiempo la llamada
metricas_operador['ineficaz_numero_llamadas_salientes'] =  (metricas_operador['outgoing_calls_mean'] < media_outgoing)

#condición 4: Evaluación del performance con respecto al tiempo la llamada
metricas_operador['ineficaz_duracion_llamadas_entrantes'] =  (metricas_operador['ingoing_calls_mean'] < media_ingoing)

metricas_operador.head()
````

````python
#Se crea una función para evaluar por cada fila si tiene alguna condición que le clasifique como ineficaz

def evaluate_performance(row):
    ineficaz_count = 0
    if row['ineficaz_perdidas']: # Si tiene en promedio más llamadas promedio que el promedio
        ineficaz_count += 1
    if row['ineficaz_tiempo_espera']: # Si en promedio el tiempo de espera para las llamadas es más alto que el de la media
        ineficaz_count += 1
    if row['ineficaz_numero_llamadas_salientes']: # Si la duración de las llamadas salientes en promedio es más alta que la de la media
        ineficaz_count += 1
    if row['ineficaz_duracion_llamadas_entrantes']: # Si la duración de las llamadas entrantes en promedio es más alta que la de la media
        ineficaz_count += 1

#Se verifica el conteo de condiciones que ha cumplido y se le califica en 4 sub categorías de eficacia.
    if ineficaz_count == 4:
        return 'Ineficaz'
    elif ineficaz_count == 3:
        return 'Parcialmente ineficaz'
    elif ineficaz_count == 2:
        return 'Parcialmente eficaz'
    else:
        return 'Eficaz'

#Se aplica la función creando una nueva columna
metricas_operador['performance_general'] = metricas_operador.apply(evaluate_performance, axis=1)

#Se comprueba que la calificación sea acorde a los parámetros establecidos
print(metricas_operador.head())

#Se traza una distribución para ver la magnitud de los resultados

plt.hist(metricas_operador['performance_general'])
plt.show()
````

Se desarrolló una función para catalogar el trabajo general de cada operador. Se crearon 4 subcategorías porque no es lo mismo ser ineficaz en una métrica que serlo en todas. La condición dada es que un operador es ineficiente si cumple al tiempo las tres (cuatro condiciones): tiempos de espera altos, muchas llamadas perdidas y muchas llamadas salientes.

1. Ineficaz: Esta subcategoría es la principal. Si hay 4 strikes se cumple con el criterio para pertener a este grupo.
2. Parcialmente ineficaz: No se cumplen todas las condiciones al tiempo, pero se cumplen 3. Es decir, si hay 3 strikes, se entra a este grupo y se está en riesgo de pasar a ser un operador ineficaz. **Esta métrica debe ser de especial atención para los supervisores**.
3. Parcialmente eficaz: Se cumplen 2 condiciones al tiempo. Es decir, si hay 2 strikes, se entra a este grupo y se está en riesgo de pasar a ser un operador parcialmente ineficaz. **Esta métrica es de especial atención para los supervisores**.
4. Eficaz: La métrica premio. Si alguien tiene un strike o menos se considera como un operador eficiente ya que no cumple varias condiciones simultaneas y un usuario disgustado puede dañar en cualquier momento sus métricas.

Se observa una distribución para entender cuál es la tendencia. Se observa que más de la mitad son  eficaces o parcialmente eficaces, una buena noticia. Sin embargo, una minoría se cuenta como ineficaz.

      H0: Son iguales las proporciones de llamadas perdidas cuando se espera mucho tiempo y cuando no.

````python
#Se realiza un filtro para tener las llamadas perdidas agrupadas por el tiempo de espera superior a la media.
 #Antes se separó por llamadas entrantes y salientes, en esta ocasión no se hará, porque este fenómeno igualmente puede aplicar a las llamadas salientes.

llamadas_tiempo_de_espera_largo = eventos_copy[eventos_copy['tiempo_espera'] > media_wait_time]
numero_llamadas_tiempo_de_espera_largo = llamadas_tiempo_de_espera_largo['internal'].count()
llamadas_rechazadas_con_tiempo_de_espera_largo = llamadas_tiempo_de_espera_largo[llamadas_tiempo_de_espera_largo['is_missed_call'] == True]['internal'].count()

llamadas_con_tiempo_de_espera_normal_o_inferior = eventos_copy[eventos_copy['tiempo_espera'] <= media_wait_time]
numero_llamadas_con_tiempo_de_espera_normal_o_inferior = llamadas_con_tiempo_de_espera_normal_o_inferior['internal'].count()
numero_llamadas_rechazadas_con_tiempo_de_espera_normal_o_inferior = llamadas_con_tiempo_de_espera_normal_o_inferior[llamadas_con_tiempo_de_espera_normal_o_inferior['is_missed_call'] == True]['internal'].count()

llamadas = eventos_copy['internal'].count()

porcentaje_llamadas_rechazadas_con_alto_tiempo_de_espera = llamadas_rechazadas_con_tiempo_de_espera_largo /llamadas

porcentaje_llamadas_rechazadas_con_de_espera_normal_o_inferior = numero_llamadas_rechazadas_con_tiempo_de_espera_normal_o_inferior /llamadas

alpha = 0.05

v1 = [llamadas_rechazadas_con_tiempo_de_espera_largo, numero_llamadas_rechazadas_con_tiempo_de_espera_normal_o_inferior]
v2= [numero_llamadas_tiempo_de_espera_largo, numero_llamadas_con_tiempo_de_espera_normal_o_inferior]
results = proportions_ztest(v1,v2)

print(results[1])
if results[1] > alpha:
  print('Las proporciones son iguales')
else:
  print('las proporciones son diferentes')

print('Porcentaje de llamadas rechazadas con alto tiempo de espera:',porcentaje_llamadas_rechazadas_con_alto_tiempo_de_espera,'Porcentaje de llamdas rechazadas con tiempo de espera inferior al normal:', porcentaje_llamadas_rechazadas_con_de_espera_normal_o_inferior )
````

Se rechaza la hipótesis nula. La proporción de llamadas rechazadas es diferente según el tiempo de espera. Esto quiere decir que la premisa es parcialmente cierta, pues el tiempo de espera si influye en el número de llamadas rechazadas, pero no es influyente. Aunque es importante ver que por pequeña que sea la diferencia influye en el número de rechazos.

Ahora, como es posible que la diferencia entre llamadas entrantes y salientes sea importante, se repetirá el procedimiento, pero solo con las llamadas entrantes, la métrica en dónde el tiempo de espera es más importante y depende de la compañía controlarlo.


````python
#Se realiza un filtro para tener las llamadas perdidas agrupadas por el tiempo de espera superior a la media.
 #En esta ocasión se hará la separación para comprobar que esta métrica es particularmente importante para las llamadas entrantes.

#dataframe filtrado por llamadas entrantes

llamadas_entrantes = eventos_copy[eventos_copy['direction']=='in']

llamadas_tiempo_de_espera_largo_in = llamadas_entrantes[llamadas_entrantes['tiempo_espera'] > media_wait_time]
numero_llamadas_tiempo_de_espera_largo_in = llamadas_tiempo_de_espera_largo_in['internal'].count()
llamadas_rechazadas_con_tiempo_de_espera_largo_in = llamadas_tiempo_de_espera_largo_in[llamadas_tiempo_de_espera_largo_in['is_missed_call'] == True]['internal'].count()


llamadas_con_tiempo_de_espera_normal_o_inferior_in = llamadas_entrantes[llamadas_entrantes['tiempo_espera'] <= media_wait_time]
numero_llamadas_con_tiempo_de_espera_normal_o_inferior_in = llamadas_con_tiempo_de_espera_normal_o_inferior_in['internal'].count()
numero_llamadas_rechazadas_con_tiempo_de_espera_normal_o_inferior_in = llamadas_con_tiempo_de_espera_normal_o_inferior_in[llamadas_con_tiempo_de_espera_normal_o_inferior_in['is_missed_call'] == True]['internal'].count()

llamadas_in = llamadas_entrantes['internal'].count()

porcentaje_llamadas_rechazadas_con_alto_tiempo_de_espera_in = llamadas_rechazadas_con_tiempo_de_espera_largo_in /llamadas_in

porcentaje_llamadas_rechazadas_con_de_espera_normal_o_inferior_in = numero_llamadas_rechazadas_con_tiempo_de_espera_normal_o_inferior_in /llamadas_in

alpha = 0.05

v1_in = [llamadas_rechazadas_con_tiempo_de_espera_largo_in, numero_llamadas_rechazadas_con_tiempo_de_espera_normal_o_inferior_in]
v2_in = [numero_llamadas_tiempo_de_espera_largo_in, numero_llamadas_con_tiempo_de_espera_normal_o_inferior_in]
results_in = proportions_ztest(v1,v2)

print(results_in[1])
if results_in[1] > alpha:
  print('Las proporciones son iguales')
else:
  print('las proporciones son diferentes')

print('Porcentaje de llamadas rechazadas con alto tiempo de espera:',porcentaje_llamadas_rechazadas_con_alto_tiempo_de_espera_in,'Porcentaje de llamdas rechazadas con tiempo de espera inferior al normal:', porcentaje_llamadas_rechazadas_con_de_espera_normal_o_inferior_in )
````

Se rechaza la hipótesis nula y se acepta la premisa: Si hay diferencias entre el número de llamadas rechazadas.
